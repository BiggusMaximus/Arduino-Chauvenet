```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv('../data/volt.csv')
data
```

# Reading Data

```python
plt.scatter(data["analog"], data["multimeter"], facecolors='none', edgecolors='k')
plt.xlabel('Analog')
plt.ylabel('Voltage')
```

```python
def eval_2nd_degree(coeffs, x):
    """
    Function to return the output of evaluating a second degree polynomial,
    given a specific x value.
    
    Args:
        coeffs: List containing the coefficients a,b, and c for the polynomial.
        x: The input x value to the polynomial.
    
    Returns:
        y: The corresponding output y value for the second degree polynomial.
    
    """
    a = (coeffs[0]*(x*x))
    b = coeffs[1]*x
    c = coeffs[2]
    y = a+b+c
    return y
```

```python

rand_coeffs = (random.randrange(-10, 10), random.randrange(-10, 10), random.randrange(-10,10))
y_bar = eval_2nd_degree(rand_coeffs, data.analog.to_list())
```

```python
plt.scatter(data["analog"].to_numpy, data["multimeter"].to_numpy(), facecolors='none', edgecolors='k', label='Original')
plt.scatter(data["analog"].to_numpy(), y_bar, facecolors='none', edgecolors='b', label='First Predict')
plt.xlabel('Analog')
plt.ylabel('Voltage')
plt.legend()
plt.show
```
